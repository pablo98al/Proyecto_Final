```python
import pandas as pd

df_rentasPV = pd.read_csv("Data_Raw/rentas_PV.csv", sep=";", encoding="utf-8-sig")
```

```python
print(df_rentasPV.shape)
```

```python
df_rentasPV
```

```python
df_rentasPV = df_rentasPV.drop(columns=["sexo"]) #Eliminamos la columan sexo
df_rentasPV
```

```python
df_rentasPV = df_rentasPV.rename(columns={"2016": "total_2016"}) #CAMBIAMOS NOMBRE COLUMNA
df_rentasPV
```

```python
df_rentasPV = df_rentasPV.pivot_table(
    index="ambitos territoriales",
    columns="tipo de renta",
    values="total_2016"
).reset_index() #Pivotamos la tabla para que se vea mas claro
```

```python
df_rentasPV
```

```python
df_rentasPV = df_rentasPV.rename(columns={"tipo de renta": "Id"})
```

```python
# Convertir todas las columnas excepto 'ambitos territoriales' a Int64
cols_int = [col for col in df_rentasPV.columns if col != "ambitos territoriales"]

for col in cols_int:
    df_rentasPV[col] = pd.to_numeric(df_rentasPV[col], errors="coerce").astype("Int64")

print(df_rentasPV.dtypes)
```

```python
df_rentasPV
```

```python
import os

# Exportar a CSV
df_rentasPV.to_csv(os.path.join("Data_Limpia/rentas_PV.csv"), index=False)

print("✅ Archivo guardado en: Pre - Proyecto/Data_Limpia/rentas_PV_2016.csv")
```
