```python
import pandas as pd

df_rentas = pd.read_csv("../Data_Raw/rentas_espanya.csv")
```

```python
#Separamos el codigo del municipio del nombre del municipio
df_rentas = df_rentas.rename(columns={"Unnamed: 0": "municipio_cod"})
df_rentas["Municipio"] = df_rentas["municipio_cod"].str.rsplit("-", n=1).str[0]
df_rentas["Id Municipio"]   = df_rentas["municipio_cod"].str.rsplit("-", n=1).str[1]
df_rentas = df_rentas.drop(columns=["municipio_cod"])
```

```python
df_rentas.head()
```

```python
#Eliminar puntos de miles y convertir las columnsa correspondientes a numérico
cols_numericas = [
    "Número de declaraciones", "Renta bruta media", "Renta disponible media",
    "Rentas del trabajo", "Rentas exentas", "Renta bruta",
    "Cotizaciones sociales a la SS", "Cuota  resultante de autoliquidación",
    "Renta disponible"
]

for col in cols_numericas:
    if df_rentas[col].dtype == object:
        df_rentas[col] = df_rentas[col].str.replace(".", "", regex=False)
    df_rentas[col] = pd.to_numeric(df_rentas[col], errors='coerce')
```

```python
df_rentas["Número de declaraciones"] = df_rentas["Número de declaraciones"].astype(int) #cambiamos a int para evitar decimales en esta columna
print(df_rentas.shape)
print(df_rentas.dtypes)
df_rentas.head()
```

```python
print(f"Duplicados: {df_rentas.duplicated().sum()}")
```

```python
# Normalizamos el nombre de la columna para que coincida con el resto de los dataframes
df_rentas = df_rentas.rename(columns={'Municipio': 'provincia'})
```
