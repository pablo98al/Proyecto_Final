```python
import pandas as pd

df_poblacion = pd.read_csv("../Data_Raw/addinfo.csv", sep=",")
```

```python
df_poblacion
```

```python
# Eliminar puntos de miles y convertir a Int64
df_poblacion["valor"] = (
    df_poblacion["valor"]
    .astype(str)
    .str.replace(".", "", regex=False)
    .pipe(pd.to_numeric, errors="coerce")
    .astype("Int64")
)

print(df_poblacion.dtypes)
print(df_poblacion.shape)
df_poblacion.head()
```

```python
split = df_poblacion["poblacion"].str.split(" ", n=1, expand=True)

df_poblacion["cod_provincia"] = split[0]  
df_poblacion["poblacion"]     = split[1]  

print(df_poblacion.head())
```

```python
df_poblacion = df_poblacion[df_poblacion["poblacion"] != "Nacional"] #Eliminamos la fila que agrupa todo
```

```python
import os

# Exportar a CSV
df_poblacion.to_csv(os.path.join("../Output/poblacion.csv"), index=False) 

print("✅ Archivo guardado en: Pre - Proyecto/Output/poblacion.csv")
```
