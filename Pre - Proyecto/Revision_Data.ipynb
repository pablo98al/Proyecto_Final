```python
import pandas as pd
import glob
import os

# Cambia esta ruta a tu carpeta
ruta = "C:\\Users\\pablo\\iCloudDrive\\Master - ThePower\\Data Analyst V3\\Repositoros_GitHub\\Proyecto_Final\\Proyecto_Final\\Pre - Proyecto\\Data - raw\\Casas"

# Busca todos los CSV que contengan 'house' en el nombre
archivos = glob.glob(os.path.join(ruta, "*house*.csv"))

print(f"Archivos encontrados: {len(archivos)}")

# Ver filas de cada archivo individualmente
for archivo in archivos:
    df_temp = pd.read_csv(archivo)
    print(f"{os.path.basename(archivo)}: {len(df_temp)} filas, {len(df_temp.columns)} columnas")

# Unir todos en un solo DataFrame
df_total = pd.concat([pd.read_csv(f) for f in archivos], ignore_index=True)

print(f"\n✅ Dataset combinado: {df_total.shape[0]} filas y {df_total.shape[1]} columnas")
```

```python
import pandas as pd
import glob
import os

# Cambia esta ruta a tu carpeta
ruta = "C:\\Users\\pablo\\iCloudDrive\\Master - ThePower\\Data Analyst V3\\Repositoros_GitHub\\Proyecto_Final\\Proyecto_Final\\Pre - Proyecto\\Data - raw"


# Busca todos los CSV que NO contengan 'house' en el nombre
todos = glob.glob(os.path.join(ruta, "*.csv"))
archivos_sin_house = [f for f in todos if "house" not in os.path.basename(f).lower()]

print(f"Archivos encontrados sin 'house': {len(archivos_sin_house)}\n")

# Ver columnas de cada archivo
for archivo in archivos_sin_house:
    df_temp = pd.read_csv(archivo, nrows=1)  # Solo leemos 1 fila para ir rápido
    print(f"📄 {os.path.basename(archivo)}:")
    print(df_temp.columns.tolist())
    print()
```
